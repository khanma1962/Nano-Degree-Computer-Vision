```python
img = cv2.imread('images/pizza_bluescreen.jpg')

print(type(img), img.shape)
plt.imshow(img)
```

```python
img_copy = np.copy(img)

img_copy = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)
plt.imshow(img_copy);
```

```python
lower_blue = np.array([0, 0, 200])
upper_blue = np.array([50, 50, 255])
```

```python
mask = cv2.inRange(img_copy, lower_blue, upper_blue)
plt.imshow(mask, cmap='gray')
```

```python
masked_img = np.copy(img_copy)
masked_img[mask != 0] = [0, 0, 0]
plt.imshow(masked_img)
```

```python
background_img = cv2.imread('images/space_background.jpg')
plt.imshow(background_img)
```

```python
background_img = cv2.cvtColor(background_img, cv2.COLOR_BGR2RGB)
cropped_background = background_img[:514, :816]

cropped_background[mask == 0] = [0, 0, 0]
plt.imshow(cropped_background)
```

```python
complete_img = masked_img + cropped_background
plt.imshow(complete_img)
```
